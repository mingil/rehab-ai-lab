```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Í∞ÄÏÉÅÏùò ÌôòÏûê Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (N=100)
np.random.seed(42) # Í≤∞Í≥º Ïû¨ÌòÑÏùÑ ÏúÑÌï¥ Í≥†Ï†ï
n_patients = 100

data = {
    'Patient_ID': range(1, n_patients + 1),
    'Age': np.random.randint(20, 80, n_patients), # 20~80ÏÑ∏
    'Treatment_Days': np.random.randint(5, 50, n_patients), # ÏπòÎ£å Í∏∞Í∞Ñ
}

# ÌöåÎ≥µ Ï†êÏàò (Recovery_Score) ÏÉùÏÑ±: ÏπòÎ£å Í∏∞Í∞ÑÏù¥ Í∏∏ÏàòÎ°ù Ï†êÏàòÍ∞Ä ÎÜíÍ≤å ÏÑ§Ï†ï (ÏïΩÍ∞ÑÏùò ÎÖ∏Ïù¥Ï¶à Ï∂îÍ∞Ä)
df = pd.DataFrame(data)
df['Recovery_Score'] = df['Treatment_Days'] * 1.5 + np.random.normal(0, 5, n_patients) + 30

# 2. Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏ (ÏÉÅÏúÑ 5Î™Ö)
print("üè• [ÌôòÏûê Îç∞Ïù¥ÌÑ∞ ÎØ∏Î¶¨Î≥¥Í∏∞]")
display(df.head())

# 3. Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî (ÎÖºÎ¨∏Ïö© Í∑∏ÎûòÌîÑ)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Treatment_Days', y='Recovery_Score', hue='Age', palette='viridis', s=100)

plt.title('Correlation between Treatment Duration and Recovery', fontsize=15)
plt.xlabel('Treatment Duration (Days)', fontsize=12)
plt.ylabel('Recovery Score (Fugl-Meyer)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 4. Í∑∏ÎûòÌîÑ Ï†ÄÏû•
plt.savefig('figure_1_recovery_analysis.png', dpi=300)
print("\n‚úÖ Í∑∏ÎûòÌîÑÍ∞Ä 'figure_1_recovery_analysis.png'Î°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")

plt.show()
```
